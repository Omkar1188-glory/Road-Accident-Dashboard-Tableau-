```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("accident data.csv")
```

```python
df.head()
```

```python
print(df.columns)
```

```python
df['Accident Date'] = pd.to_datetime(df['Accident Date'], errors='coerce')
df['Accident Date'] = df['Accident Date'] + pd.DateOffset(years=3)
```

```python
df.head(100)
```

```python
print(df['Accident Date'].min())
print(df['Accident Date'].max())
```

```python
df.to_csv("accident_data_2022_2025.csv", index=False)
```

```python
# Sort by index or another stable column
df = df.sort_index()

# Fill missing dates
df['Accident Date'] = df['Accident Date'].ffill().bfill()
```

```python
df.to_csv("accident_data_.csv", index=False)
```
